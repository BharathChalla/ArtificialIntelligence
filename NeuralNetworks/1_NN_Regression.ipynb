#Linear Regression on Boston Housing Dataset

##Import the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

%matplotlib inline
```

###Boston Housing Dataset

```python
from sklearn.datasets import load_boston
boston_dataset = load_boston()
# import pandas as pd
# import numpy as np


# data_url = "http://lib.stat.cmu.edu/datasets/boston"
# raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
# data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
# boston_dataset = raw_df.values[1::2, 2]
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston.head()
```

```python
boston['MEDV'] = boston_dataset.target
```

##Preparing the data for training the model

```python
from sklearn.preprocessing import StandardScaler

target_col = 'MEDV'
X = boston.loc[:, boston.columns != target_col]
Y = pd.DataFrame(boston[target_col])

scaler = StandardScaler()
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X), columns=boston_dataset.feature_names)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

##Training and testing the model

```python
import torch
import torch.nn as nn

device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
class RegressionNN(nn.Module):

    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.hidden_layer1 = nn.Linear(input_dim, 16)
        self.hidden_layer2 = nn.Linear(16, 32)
        self.output = nn.Linear(32, output_dim)

    def forward(self, x):
        x = torch.relu(self.hidden_layer1(x))
        x = torch.tanh(self.hidden_layer2(x))
        x = self.output(x)
        return x

regression_nn_model = RegressionNN(X_train.shape[1], 1)
print(regression_nn_model)
```

```python
X_train
```

```python
def train(X_train, Y_train, model, epochs=50000, verbose=True):
    X = torch.tensor(X_train.values, dtype=torch.float)
    y = torch.tensor(Y_train.values, dtype=torch.float)
    loss_function = nn.L1Loss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
    for epoch in range(epochs):
        optimizer.zero_grad()
        output = model(X)
        loss = loss_function(output, y)
        loss.backward()
        optimizer.step()
        if(verbose and epoch % (epochs/10) == 0):
            print(("Epoch :{0:>6} ----> Training Loss: {1}").format(str(epoch), str(loss.item())))
    return model
```

```python
print(X_train.shape)
print(Y_train.shape)
trained_model = train(X_train, Y_train, regression_nn_model)
```

##Model evaluation

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# model evaluation for training set
X_tensor = torch.tensor(X_train.values, dtype=torch.float)
y_train_predict = trained_model(X_tensor)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict.detach().numpy())))
r2 = r2_score(Y_train, y_train_predict.detach().numpy())

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
X_tensor = torch.tensor(X_test.values, dtype=torch.float)
y_test_predict = trained_model(X_tensor)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict.detach().numpy())))
r2 = r2_score(Y_test, y_test_predict.detach().numpy())

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
